# Data Cleaning

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_train = pd.read_csv(r'./west_nile/input/train.csv')
```

```python
df_train.head()
```

```python
df_train.info()
```

We can see that there are no null values from the info.

From the data, we will only keep the Data, Species, Trap, Latitude, Longitude, NumMosquitos, WnvPresent.

We will typecast them into their appropriate datatype.

```python
(df_train
 [['Date','Species','Trap','Latitude','Longitude','NumMosquitos','WnvPresent']]
 .assign(Date = pd.to_datetime(df_train.Date))
 .astype({'Species':'category', 'Trap':'category','NumMosquitos':'int8','WnvPresent':'bool'})
)
```

From the way the data was collected, we know that if there are over 50 mosquitos, they will create another entry.

```python
(df_train[df_train.duplicated(['Date','Trap'],keep=False)]
 .sort_values(by=['NumMosquitos','Date','Trap'], ascending=False)
 .head(20)
)
```

We will assume that all duplicate entries are either data collected at different times of the day or an additional entry to that was segregated.

Also we will assume that if West Nile Virus is found in one of the entries, then it is present in that area (i.e. we will change all `WnvPresent` to `True` if at least one of them if `True`.

We will only add them together if the species are the same as we'd like to see which species 

```python
(df_train
 [['Date','Species','Trap','Latitude','Longitude','NumMosquitos','WnvPresent']]
 .assign(Date = pd.to_datetime(df_train.Date))
 .astype({'Species':'category', 'Trap':'category','NumMosquitos':'int8','WnvPresent':'bool'})
 .groupby(by=['Date','Species','Trap'])
 .agg({
      'Latitude':np.mean,
      'Longitude':np.mean,
      'NumMosquitos':np.sum,
      'WnvPresent':np.max})
 .dropna()
 .sort_values(['Date','Trap'])
 .reset_index()
)
```

```python
(df_train
 [['Date','Species','Trap','Latitude','Longitude','NumMosquitos','WnvPresent']]
 .assign(Date = pd.to_datetime(df_train.Date))
 .astype({'Species':'category', 'Trap':'category','NumMosquitos':'int8','WnvPresent':'bool'})
 .groupby(by=['Date','Species','Trap'])
 .agg({
      'Latitude':np.mean,
      'Longitude':np.mean,
      'NumMosquitos':np.sum,
      'WnvPresent':np.max})
 .dropna()
 .sort_values(['NumMosquitos'])
 .reset_index()
)
```

Since the number of mosquitos is over 127, we will cast NumMosquitos as int16.

After processing using np.max, WnvPresent may not be boolean so we will typecast it again.

```python
(df_train
 [['Date','Species','Trap','Latitude','Longitude','NumMosquitos','WnvPresent']]
 .assign(Date = pd.to_datetime(df_train.Date))
 .astype({'Species':'category', 'Trap':'category','NumMosquitos':'int8','WnvPresent':'bool'})
 .groupby(by=['Date','Species','Trap'])
 .agg({
      'Latitude':np.mean,
      'Longitude':np.mean,
      'NumMosquitos':np.sum,
      'WnvPresent':np.max})
 .dropna()
 .sort_values(['Date','Trap'])
 .reset_index()
 .astype({'NumMosquitos':'int16','WnvPresent':'boolean'})
).info()
```

We will put the cleaned data into a function so our team members can replicate the dataframe.

```python
def clean_train(train_data):
    return (train_data
 [['Date','Species','Trap','Latitude','Longitude','NumMosquitos','WnvPresent']]
 .assign(Date = pd.to_datetime(df_train.Date))
 .astype({'Species':'category', 'Trap':'category','NumMosquitos':'int8','WnvPresent':'bool'})
 .groupby(by=['Date','Species','Trap'])
 .agg({
      'Latitude':np.mean,
      'Longitude':np.mean,
      'NumMosquitos':np.sum,
      'WnvPresent':np.max})
 .dropna()
 .sort_values(['Date','Trap'])
 .reset_index()
 .astype({'NumMosquitos':'int16','WnvPresent':'boolean'})
)
```

```python
df = clean_train(df_train)
```

```python
df.info()
```

# Weather Data

```python
df_weather = pd.read_csv(r'./west_nile/input/weather.csv')
df_weather.shape
```

```python
df_weather['Heat'].value_counts()
```

```python
df_weather.info()
```

From the kaggle information, it is believed that hot and dry conditions are more favorable for WNV than cold and wet.

We will take a subset of features and merge with our main data frame after cleaning.

Since two weather stations are given, we might want to impute any missing values based on which station has the information. We will add in the Lat and Long of the weather stations.

Station 1: (Lat 41.995,Long -87.933), Station 2: (Lat 41.786,Long -87.752)

We will also take all the temperature measurements to explore the conditions for the various years that we have. Note that Depart is the feature of the temperature departure from normal.

The dew point is the temperature to which air must be cooled to become saturated with water vapor, assuming constant air pressure and water content. This is a good measure of how wet surfaces will be which can be used in conjunction with `Tmin`.

The wet bulb temperature is always lower than the dry bulb temperature but will be identical with 100% relative humidity (the air is at the saturation line).

The features `HEATING` and `COOLING` are a measure of Degree Days.
Degree days are the difference between the daily temperature mean, (high temperature plus low temperature divided by two) and 65°F. So it shows how much a building needs to be heated or cooled to achieve the base of 65°F. Thus it is an indicator of the temperature. We will leave these features out since we already have temperature data.

`Sunrise` and `Sunset` are not directly related to weather conditions.  So we will leave those out.

`CodeSum` gives us Significant Weather Types. We might be interested in the following

|Weather|Hot and Dry|Wet and Cold|
|---|---|---|
||HZ,FU,VA,DU,DS,PO,SA,SS|+FC, FC, TS, RA,DZ, BR,PY, SQ,SH,MI,PR|

13,14,15 are regarding Snow/Ice so it is not important for us.

Precipitation is any liquid or frozen water that forms in the atmosphere and falls back to the earth. It would be useful to know since it is an indication of wet weather.

We will leave wind data out because we can infer from `CodeSum`.

```python
(df_weather
 [['Station','Date','Tmax','Tmin','Tavg', 'Depart', 'DewPoint','CodeSum','PrecipTotal']]
 .assign(Date=pd.to_datetime(df_weather['Date']),
     Latitude=df_weather['Station'].map(lambda x:41.995 if x==1 else 41.786),
     Longitude=df_weather['Station'].map(lambda x:-87.933 if x==1 else -87.752))
 .astype({'Station':'category','Tmax':'int8','Tmin':'int8'})
)[['Station','Date','Tavg']].head()
```

```python
(df_weather
 [['Station','Date','Tmax','Tmin','Tavg', 'Depart', 'DewPoint','CodeSum','PrecipTotal']]
 .assign(Date=pd.to_datetime(df_weather['Date']),
     Latitude=df_weather['Station'].map(lambda x:41.995 if x==1 else 41.786),
     Longitude=df_weather['Station'].map(lambda x:-87.933 if x==1 else -87.752))
 .astype({'Station':'category','Tmax':'int8','Tmin':'int8'})
 [['Station','Tavg']]
 .query('Tavg=="M"')
)
```

We will impute all the missing values with Station 1's average temperature.

```python
(df_weather
 [['Station','Date','Tmax','Tmin','Tavg', 'Depart', 'DewPoint','CodeSum','PrecipTotal']]
 .assign(Date=pd.to_datetime(df_weather['Date']),
     Latitude=df_weather['Station'].map(lambda x:41.995 if x==1 else 41.786),
     Longitude=df_weather['Station'].map(lambda x:-87.933 if x==1 else -87.752),
     Tavg=df_weather['Tavg'].map(lambda x: np.nan if x=='M' else x).ffill())
 .astype({'Station':'category','Tmax':'int8','Tmin':'int8'})
)
```

```python
(df_weather
 [['Station','Date','Tmax','Tmin','Tavg', 'Depart', 'DewPoint','CodeSum','PrecipTotal']]
 .assign(Date=pd.to_datetime(df_weather['Date']),
     Latitude=df_weather['Station'].map(lambda x:41.995 if x==1 else 41.786),
     Longitude=df_weather['Station'].map(lambda x:-87.933 if x==1 else -87.752))
 .astype({'Station':'category','Tmax':'int8','Tmin':'int8'})
 [['Station','Depart']]
 .query('Depart=="M"')
).value_counts()
```

We will forward fill with Station 1 data.

```python
(df_weather
 [['Station','Date','Tmax','Tmin','Tavg', 'Depart', 'DewPoint','CodeSum','PrecipTotal']]
 .assign(Date=pd.to_datetime(df_weather['Date']),
     Latitude=df_weather['Station'].map(lambda x:41.995 if x==1 else 41.786),
     Longitude=df_weather['Station'].map(lambda x:-87.933 if x==1 else -87.752),
     Tavg=df_weather['Tavg'].map(lambda x: np.nan if x=='M' else x).ffill(),
     Depart=df_weather['Depart'].map(lambda x: np.nan if x=='M' else x).ffill())
 .astype({'Station':'category','Tmax':'int8','Tmin':'int8','Tavg':'int8','Depart':'int8'})
).describe()
```

Since they are all station 2, we will also forward fill with Station 1's data.

```python
(df_weather
 [['Station','Date','Tmax','Tmin','Tavg', 'Depart', 'DewPoint','CodeSum','PrecipTotal']]
 .assign(Date=pd.to_datetime(df_weather['Date']),
     Latitude=df_weather['Station'].map(lambda x:41.995 if x==1 else 41.786),
     Longitude=df_weather['Station'].map(lambda x:-87.933 if x==1 else -87.752),
     Tavg=df_weather['Tavg'].map(lambda x: np.nan if x=='M' else x).ffill(),
     Depart=df_weather['Depart'].map(lambda x: np.nan if x=='M' else x).ffill())
 .astype({'Station':'category','Tmax':'int8','Tmin':'int8','Tavg':'int8','Depart':'int8'})
)['PrecipTotal'].value_counts()
```

Since T stands for Trace amounts that are too small, we will impute the value 0.

```python
(df_weather
 [['Station','Date','Tmax','Tmin','Tavg', 'Depart', 'DewPoint','CodeSum','PrecipTotal']]
 .assign(Date=pd.to_datetime(df_weather['Date']),
     Latitude=df_weather['Station'].map(lambda x:41.995 if x==1 else 41.786),
     Longitude=df_weather['Station'].map(lambda x:-87.933 if x==1 else -87.752),
     Tavg=df_weather['Tavg'].map(lambda x: np.nan if x=='M' else x).ffill(),
     Depart=df_weather['Depart'].map(lambda x: np.nan if x=='M' else x).ffill(),
     PrecipTotal=df_weather['PrecipTotal'].map(lambda x: 0 if x=='  T' else x))
 .astype({'Station':'category','Tmax':'int8','Tmin':'int8','Tavg':'int8','Depart':'int8'})
)[df_weather['PrecipTotal']=='M']
```

Again, we will impute the values from Station 1.

```python
(df_weather
 [['Station','Date','Tmax','Tmin','Tavg', 'Depart', 'DewPoint','CodeSum','PrecipTotal']]
 .assign(Date=pd.to_datetime(df_weather['Date']),
     Latitude=df_weather['Station'].map(lambda x:41.995 if x==1 else 41.786),
     Longitude=df_weather['Station'].map(lambda x:-87.933 if x==1 else -87.752),
     Tavg=df_weather['Tavg'].map(lambda x: np.nan if x=='M' else x).ffill(),
     Depart=df_weather['Depart'].map(lambda x: np.nan if x=='M' else x).ffill(),
     PrecipTotal=df_weather['PrecipTotal'].map(lambda x: 0 if x=='  T' else x).map(lambda x: np.nan if x=='M' else x).ffill())
 .astype({'Station':'category','Tmax':'int8','Tmin':'int8','Tavg':'int8','Depart':'int8','PrecipTotal':'float32'})
 .join(df_weather['CodeSum'].str.get_dummies(sep=' '))
 .drop('CodeSum', axis = 1)
)['DewPoint'].value_counts()
```

```python
def weather(df_weather):
    return (df_weather
     [['Station','Date','Tmax','Tmin','Tavg', 'Depart', 'DewPoint','CodeSum','PrecipTotal']]
     .assign(Date=pd.to_datetime(df_weather['Date']),
         Latitude=df_weather['Station'].map(lambda x:41.995 if x==1 else 41.786),
         Longitude=df_weather['Station'].map(lambda x:-87.933 if x==1 else -87.752),
         Tavg=df_weather['Tavg'].map(lambda x: np.nan if x=='M' else x).ffill(),
         Depart=df_weather['Depart'].map(lambda x: np.nan if x=='M' else x).ffill(),
         PrecipTotal=df_weather['PrecipTotal'].map(lambda x: 0 if x=='  T' else x).map(lambda x: np.nan if x=='M' else x).ffill())
     .astype({'Station':'category','Tmax':'int8','Tmin':'int8','Tavg':'int8','Depart':'int8','PrecipTotal':'float32', 'DewPoint':'int8'})
     .join(df_weather['CodeSum'].str.get_dummies(sep=' '))
     .drop('CodeSum', axis = 1)
    )
```

```python
df_weather = weather(df_weather)
```
